```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Deleting Reduntant Columns

```python
col_to_del = ['Attrition','Over18', 'EmployeeCount', 'StandardHours']
hr_data.drop(col_to_del, axis = 1, inplace = True)
```

```python
hr_data.head()
```

Deleting Duplicates

```python
# Count rows before deleting duplicates
rows_before = len(hr_data)

# Remove duplicates and keep the first occurrence
hr_data_no_duplicates = hr_data.drop_duplicates()

# Count rows after deleting duplicates
rows_after = len(hr_data_no_duplicates)

print(f"Rows before: {rows_before}")
print(f"Rows after: {rows_after}")
```

Correlation map for all numeric variables

```python
numeric_columns = hr_data.select_dtypes(include=['number'])
correlation_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(17, 15))
sns.heatmap(correlation_matrix, annot=True, cmap="seismic", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Create a pie for Overtime
Overtime_counts = hr_data['OverTime'].value_counts()
plt.pie(Overtime_counts, labels=Overtime_counts.index, autopct='%1.1f%%', colors=['skyblue', 'darkblue'])
plt.title('Pie Chart for a OverTime')
plt.show()
```

```python
# Marital Status of employees
status_counts = hr_data['MaritalStatus'].value_counts()

plt.figure(figsize=(6, 4)),
plt.bar(status_counts.index, status_counts.values, color='darkblue'),
plt.xlabel('MaritalStatus'),
plt.ylabel('Values'),
plt.title('Distribution of Marital Status'),
plt.show()
```

```python
# Roles of employees
jobrole_counts = hr_data['JobRole'].value_counts()

plt.figure(figsize=(6, 4))
plt.barh(jobrole_counts.index, jobrole_counts.values, color='darkblue')
plt.xlabel('No. of employees')
plt.title('Distribution of Job Role')
plt.show()
```

```python
# Create a pie for gender
gender_counts = hr_data['Gender'].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['darkblue', 'skyblue'])
plt.title('Distribution of Gender')
plt.show()
```

```python
# Field of study

educationfield_counts = hr_data['EducationField'].value_counts()

plt.figure(figsize=(6, 4))
plt.barh(educationfield_counts.index, educationfield_counts.values, color='darkblue')
plt.xlabel('Number of Employees')
plt.title('Distribution of Education Field')
plt.show()
```

```python
#Visualization to ascertain number of employees in the different department
dept_counts = hr_data['Department'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(dept_counts.index, dept_counts.values, color='darkblue')
plt.ylabel('Number of Employees')
plt.title('Distribution of Department')
plt.show()
```

```python
#Graph showing copany sponsored travels

biz_travels_counts = hr_data['BusinessTravel'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(biz_travels_counts.index, biz_travels_counts.values, color='darkblue')
plt.ylabel('Number of Employees')
plt.title('Distribution of Business Travel')
plt.show()
```

```python
# Create a grouped bar chart to check the relationship between overtime and age
plt.figure(figsize=(8, 6))
sns.boxplot(data=hr_data, x='OverTime', y='Age', palette='Set1')
plt.title('Relationship Between Overtime and Age')
plt.show()
```

```python
#Number of years an employee has worked
plt.figure(figsize=(6, 4))
plt.hist(hr_data['TotalWorkingYears'],bins=10, color='darkblue', edgecolor='black')
plt.xlabel('Sum of working Years')
plt.ylabel('Frequency')
plt.title('Total Working Years')
plt.show()
```

```python
# Education level of the employees 1 being basic
edu_counts = hr_data['Education'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(edu_counts.index, edu_counts.values, color='darkblue')
plt.xlabel('Levels')
plt.ylabel('Number of Employees')
plt.title('Education Level')
plt.show()
```

```python
# Number of companies employee worked
no_coy_worked = hr_data['NumCompaniesWorked'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(no_coy_worked.index, no_coy_worked.values, color='darkblue')
plt.xlabel('No. of Companies')
plt.ylabel('Number of Employees')
plt.title('Number of Companies Worked')
plt.show()
```

```python
#Visualization showing distance from home
plt.figure(figsize=(6, 4))
plt.hist(hr_data['DistanceFromHome'],bins=15, color='darkblue', edgecolor='black')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.title('Distance From Home')
plt.show()
```
